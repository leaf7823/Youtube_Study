```
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```
train_df = pd.read_csv('train.csv')
test_df  = pd.read_csv('test.csv')
```

```
train_df.head()
```

```
test_df.head()
```

```
# train testの統合

all_df = pd.concat((train_df.loc[:, 'Pclass':'Embarked'],test_df.loc[:, 'Pclass':'Embarked']))

all_df.info()
```

```
# 欠損値の埋め

all_df['Age'] = all_df['Age'].fillna(all_df['Age'].mean())   #mean 平均値
all_df['Fare'] = all_df['Fare'].fillna(all_df['Fare'].mean()) 
all_df['Embarked'] = all_df['Embarked'].fillna(all_df['Embarked'].mode()[0]) 
```

```
all_df.info()
```

```
# lebel エンコード適用　sex、embarkedを数値に変換

from sklearn.preprocessing import LabelEncoder

cat_features  = ['Sex', 'Embarked']

for col in cat_features:
    lb1 = LabelEncoder()
    all_df[col] = lb1.fit_transform(list(all_df[col].values))
```

```
all_df = all_df.drop(columns=['Name', 'Ticket', 'Cabin'])
```

```
all_df.head()
```

```
# ファイル分割

train = all_df[:train_df.shape[0]]
test  = all_df[train_df.shape[0]:]
train.info()
```

```
# 正解ラベルの抽出

y = train_df['Survived']
ID = test_df['PassengerId']
```

```
# ファイル分割

X_train, X_test ,y_train, y_test = train_test_split(train, y,random_state=0)
```

```
import xgboost as xgb
```

```
# xgboost parameter

params= {
    "objective": "binary:logistic",
    "eval_metric": "auc",
    "eta": 0.1,
    "max_depth":6,
    "subsample":1,
    "colsample_bytree": 1,
    "silent": 1
}

dtrain = xgb.DMatrix(X_train, label=y_train)
dtest = xgb.DMatrix(X_test, label=y_test)

model = xgb.train(params=params,
                  dtrain=dtrain,
                  num_boost_round=100,
                  early_stopping_rounds=10,
                  evals=[(dtest, 'aaat')]
                 )
```

```
prediction = model.predict(xgb.DMatrix(test), ntree_limit=model.best_ntree_limit)
prediction
```

```
prediction = np.where(prediction < 0.5, 0, 1)
prediction
```

```
submission = pd.DataFrame({
    'passengerId': ID,
    'Survived': prediction
})

submission.to_csv('submission.csv', index=False)
```
